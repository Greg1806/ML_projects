### Market Basket Analysis (MBA) - Support, Confidence, Lift - Apriori Algo

* https://towardsdatascience.com/data-mining-market-basket-analysis-with-apriori-algorithm-970ff256a92c 
* https://medium.com/@jihargifari/how-to-perform-market-basket-analysis-in-python-bd00b745b106 
<br> “which products should be discounted? How do we apply the discounts? Are there any patterns that occur to our sales, so that we could apply the promotion more accurately?” 
* https://medium.com/analytics-vidhya/boosting-telco-marketing-capabilities-using-market-basket-analysis-1ade1d67a601 
<br> “Apart from the obvious profit from extra products sold, It also increases the dependence of the customer on the vendor and therefore reduces churn by increasing stickiness. 
<br> This is especially important in the area of telecommunications, characterized by high volatility and low customer loyalty. Customer loyalty is quite low in this sector, mostly due to anti-monopoly measures taken by governments, as well as lucrative offers for new customers from other service providers.“ 

**Note:** Potentially less affected by MBA rules as typcially go into supermarket with a list already ready (not a lot of impulsive buying :) )

**Definitions**
* **Support:** basic probability of an event happening => Chances that the product is in the transaction
* **Confidence:** probability of event A (consequent) happening given that event B (antecedent) has already happened (how often the rule has been proven to be true) => divide number of transactions with both products by transaction with single product
* **Lift:** how likely an item is purchased when another item is purchased, while controlling for how popular both items are => Lift(A -> B) = Support(A,B) / Support(A) x Support(B)
* If an itemset satisfies minimum support,then it is a frequent itemset.
* Rules that satisfy both a minimum support threshold and a minimum confidence threshold are called **Strong Association rules**

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from sklearn.preprocessing import OneHotEncoder # use pd.get_dummies() instead
```

```python
df_agg.head()
```

```python
pd.get_dummies(df_agg[['purchase_id','Artikel']], columns=['Artikel']).groupby('purchase_id').sum().max().sort_values(ascending=False).head(10)
```

```python
# data pre-processing to match Apriori algo input format
df_agg_mba = pd.get_dummies(df_agg[['purchase_id','Artikel']], columns=['Artikel']).groupby('purchase_id').sum() > 0
df_agg_mba.head(3)
```

```python
frequent_itemsets = apriori(df_agg_mba, min_support= 6/df_agg_mba.shape[0], use_colnames=True, max_len = 2)
rules = association_rules(frequent_itemsets, metric="lift",  min_threshold = 1.5)
print("Rules identified: ", len(rules))
rules.sort_values(by='support', ascending=False).head()
```

```python
rules.sort_values(by='confidence', ascending=False).head()
```

```python
rules.sort_values(by='lift', ascending=False).head()
```

```python
sns.scatterplot(data=rules, x='confidence', y='lift', size='support', sizes=(5,200))
plt.title('Association rules using MBA - 2021 & 2022 data')
plt.show()
```

```python
rules[(rules['confidence'] >= 0.5) & (rules['lift'] >= 5)].sort_values(by='support', ascending=False).head(10)
# Probably good examples: Farine -> levure OR Pate gateau -> Epinards haché OR Mozza -> Tomates Grappe

# Interpretation below
# 50%+ of transaction with “Epinards" also had “Pate à gateau”
# It is 5.5 times more likely that a I will purchase “Pate a gateau” if I have purchased “Epinards”
```
